## 匯入資料

```python
import pandas as pd
import numpy as np
#df_org=pd.read_csv(filepath+"application_train_merge_clear_nomissing .csv")
df_org=pd.read_csv("application_train_merge_clear_nomissing_with_ylabel.csv")
```

```python
df=df_org
```

## 資料切割

```python
df_X = df[['DAYS_LAST_PHONE_CHANGE', 'DAYS_LATE_PAYMENT', 'PC_SK_DPDMix_Max',
       'SK_ID_PREV_y', 'bureau_DAYS_CREDIT']]
df_y = df['simple_labelx']
```

```python
from sklearn.model_selection import train_test_split
#X , y = df_X.iloc[257511:,:] , df_y.iloc[257511:]
X , y = df_X  , df_y
X_train, X_test, y_train, y_test =\
    train_test_split(X, y, 
                     test_size=0.2,#超過10萬9,1分比較好 
                     random_state=0, 
                     stratify=y)
```

##　KNN 8/2分

```python
from sklearn.neighbors import KNeighborsClassifier
k = 10
knn=KNeighborsClassifier(n_neighbors=k)  
knn.fit(X_train, y_train)  
```

```python
print('----KNN模式訓練後，取test data 進行分類的準確率計算-------')
print('準確率:',knn.score(X_test,y_test))
```

```python
s = []
for i in range(3,11):
  k=i
  knn=KNeighborsClassifier(n_neighbors=k)  
  knn.fit(X_train, y_train)  # 用 training data 去訓練模型
  print('k =',k,' 準確率:',knn.score(X_test,y_test))  #用 test data 檢測模型的準確率
  s.append(knn.score(X_test,y_test))
```

```python
k = 10
knn=KNeighborsClassifier(n_neighbors=k)  
knn.fit(X_train, y_train)
```

```python
print('分類的預測結果：')
pred = knn.predict(X_test) 
print(pred) #觀察預測結果AA
print('真實數據：')
print(y_test.values)  #觀察真實數據(Test data)
```

```python
len(pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred)
```
